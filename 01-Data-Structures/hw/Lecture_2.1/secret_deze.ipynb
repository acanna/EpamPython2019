```python
import matplotlib.pyplot as plt
from matplotlib_venn_wordcloud import venn2_wordcloud
```

```python
pizza = {"dough", "tomatoes", "pepperoni", "ground pepper", "sweet basil",
         "a lot of cheeeese", "onion", "garlic", "salt", "oregano"}
shaverma = {"lavash", "cucumbers", "tomatoes", "sauce", "fried chicken", "onion", "cabbage"}
```

```python
venn2_wordcloud([pizza, shaverma])
```

```python
pizza.isdisjoint(shaverma)
```

```python
pizza.issubset(shaverma), shaverma.issubset(pizza)
```

```python
shaverma.issuperset(pizza), pizza.issuperset(shaverma)
```

```python
pizza & shaverma
```

```python
pizza | shaverma
```

```python
pizza - shaverma
```

```python
shaverma - pizza
```

```python
shaverma ^ pizza
```

```python
shaverma ^ pizza == (shaverma - pizza) | (pizza - shaverma)
```

```python
pizza.difference(shaverma) == pizza - shaverma
```
